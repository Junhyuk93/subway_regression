#### 역별 500m내에 존재하는 버스정류소들을 딕셔너리에 담아둠

```python
from haversine import haversine
```

```python
subloc = (sub_xy.x[0], sub_xy.y[0])
busloc = (bus_xy.x[0], bus_xy.y[0])
# haversine(latitude,longtitude) ,위도 경도 (37...,126...)
haversine(subloc, busloc, unit='m')
```

```python
dic = {}

for i in range(len(sub_xy)):
    l = []
    for j in range(len(bus_xy)):
        subloc = (sub_xy.x[i], sub_xy.y[i])
        busloc = (bus_xy.x[j], bus_xy.y[j])
        hav = haversine(subloc, busloc, unit='m')
        if hav <= 500:
            l.append(bus_xy.iloc[j, 0])
    dic[sub_xy.iloc[i, 0]] = l
```
